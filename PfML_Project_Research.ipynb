**DATA VISUALIZATION**

**Input**

Firstly, delete unnecessary parameters

```python
input_dataset = data.drop(['track_name',
                         'artist(s)_name',
                         'artist_count',
                         'released_year',
                         'released_month',
                         'in_spotify_playlists',
                         'in_spotify_charts',
                         'streams',
                         'in_apple_playlists',
                         'in_apple_charts',
                         'in_deezer_playlists',
                         'in_deezer_charts',
                         'in_shazam_charts',], axis=1)


input_dataset.head()
```

And display heatmap of inputs

```python
from sklearn.preprocessing import LabelEncoder, LabelBinarizer

le = LabelEncoder()
le.fit(input_dataset['key'])
input_dataset['key'] = le.transform(input_dataset['key'])

lb = LabelBinarizer()
lb.fit(input_dataset['mode'])
input_dataset['mode'] = lb.transform(input_dataset['mode'])


input_dataset.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = input_dataset.corr()
plt.figure(figsize=(9,9))
sns.heatmap(corr, square=True, annot=True)
```

As we can see, for choosen parameters can be observed stronger correlations for:

*   (!) danceability - valence
*   danceability - energy
*   danceability - acousticness
*   (!) valence - energy
*   (!) energy - acousticness

Because of that we get rid of:
*   valene
*   energy

```python
input_dataset = input_dataset.drop(['valence_%',
                                     'energy_%'], axis=1)


input_dataset.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = input_dataset.corr()
plt.figure(figsize=(9,9))
sns.heatmap(corr, square=True, annot=True)
```
